```python
import pandas as pd
subscriber['yearMonth'] =subscriber.account_creation_date.apply(lambda x: pd.to_datetime(x).strftime('%Y%m'))
```

```python
subscriber
```

```python
num_bing =subscriber[subscriber['attribution_technical']=='bing'].groupby('yearMonth').count()['attribution_technical']
num_display=subscriber[subscriber['attribution_technical']=='display'].groupby('yearMonth').count()['attribution_technical']
num_youtube=subscriber[subscriber['attribution_technical']=='youtube'].groupby('yearMonth').count()['attribution_technical']
num_facebook=subscriber[subscriber['attribution_technical']=='facebook'].groupby('yearMonth').count()['attribution_technical']
num_search=subscriber[subscriber['attribution_technical']=='search'].groupby('yearMonth').count()['attribution_technical']
```

```python
num_bing
```

```python
#Calculate bing CAC
sub_bing = list(num_bing)
df_bing['num_bing']= sub_bing[1:9]
df_bing['CAC']= df_bing['num_bing']/df_bing['spend_AED']
df_bing
```

```python
#Calculate display CAC
sub_display = list(num_display)
df_display['num_display']= sub_display[1:9]
df_display['CAC']= df_display['num_display']/df_display['spend_AED']
df_display
```

```python
#Calculate facebook CAC
sub_facebook = list(num_facebook)
df_facebook['num_facebook']= sub_facebook[1:9]
df_facebook['CAC']= df_facebook['num_facebook']/df_facebook['spend_AED']
df_facebook
```

```python
#Calculate youtube CAC
sub_youtube  = list(num_youtube)
df_youtube['num_youtube']= sub_youtube [1:9]
df_youtube['CAC']= df_youtube['num_youtube']/df_youtube['spend_AED']
df_youtube
```
